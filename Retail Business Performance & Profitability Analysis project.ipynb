```python
df = pd.read_csv('Sample - Superstore.csv', encoding='ISO-8859-1')
df.dropna(inplace=True)

# Convert Order Date
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
```

```python
df['Profit_Margin'] = df['Profit'] / df['Sales']
df['Profit_Margin'] = df['Profit_Margin'].replace([float('inf'), -float('inf')], 0).fillna(0)
```

```python
category_profit = df.groupby(['Category', 'Sub-Category'])[['Sales', 'Profit', 'Profit_Margin']].mean().reset_index()
category_profit.sort_values(by='Profit_Margin', ascending=True)
```

```python
import sqlite3

# Create SQLite connection
conn = sqlite3.connect(":memory:")
df.to_sql("Superstore", conn, index=False)

# Run SQL query
query = """
SELECT
  Category,
  [Sub-Category],
  AVG(Sales) AS Avg_Sales,
  AVG(Profit) AS Avg_Profit,
  AVG(Profit) / NULLIF(AVG(Sales), 0) AS Profit_Margin
FROM
  Superstore
GROUP BY
  Category, [Sub-Category]
ORDER BY
  Profit_Margin ASC;
"""
sql_result = pd.read_sql_query(query, conn)
sql_result.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.barplot(data=category_profit, x='Sub-Category', y='Profit_Margin', hue='Category')
plt.xticks(rotation=45)
plt.title("Profit Margin by Sub-Category")
plt.tight_layout()
plt.show()
```

```python
correlation = df[['Quantity', 'Profit_Margin']].corr()

plt.figure(figsize=(5, 4))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation: Quantity vs Profit Margin")
plt.show()
```

```python
monthly_sales = df.groupby(['Year', 'Month'])['Sales'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_sales, x='Month', y='Sales', hue='Year', marker='o')
plt.title("Monthly Sales Trend")
plt.xticks(range(1, 13))
plt.grid(True)
plt.tight_layout()
plt.show()
```
